```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Read the Dataset
data = pd.read_csv('cost-revenue-clean.csv')
X = data.iloc[:,:-1].values
Y = data.iloc[:,1].values
```

```python
# Visualizing The Dataset (Scatter Plot)
plt.figure(figsize=(10,6))
plt.scatter(X,Y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
# Train, Test & Split the Dataset
x_train, x_test,y_train,y_test = train_test_split(X,Y,test_size =0.2)
regression = LinearRegression()
```

```python
# Train the model, predict & Score 
regression.fit(x_train,y_train)
predict = regression.predict(x_test)
regression.score(x_test,y_test)
```

```python
# Plotting the Graph with Regression Line
plt.figure(figsize=(10,6))
plt.scatter(x_test,y_test, alpha=0.3)
plt.plot(x_test,predict, color='red', linewidth=4)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,2100000000)
plt.xlim(0,330000000)
plt.show()
```
